# **Import some library of the Python that is used in our Project for Iris Dataset Analysis**

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Import the Iris Dataset**

```python
# Load the dataset
file_path = "Iris_dataset.csv"
df = pd.read_csv(file_path)
```

# **Perform Exploratory Data Analysis in our Data**

1. Display the first five row in our data

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())
```

2. Derive the summary statistics of our Data

```python
# Summary statistics
print("\nSummary statistics:")
print(df.describe())
```

3. Check the missing value is present in our Dataset or not

```python
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

4. Show the Datatypes of our dataset

```python
# Data types of each column
print("\nData types of each column:")
print(df.dtypes)
```

5. Distribute the species in our Dataset

```python
# Distribution of species
print("\nSpecies distribution:")
print(df['species'].value_counts())
```

# **Create the Charts For Visualizing the Iris Dataset**

1. Create the Pairplot for Visualize the Relationship

```python
# Pairplot to visualize relationships
print("\nCreating pairplot for relationships...")
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.show()
```

2. Create a Heatmap for Correlation

```python
# Heatmap for correlation
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

3. Creating Box plot for feature distribution by Species

```python
# Boxplot for feature distributions by species
print("\nCreating boxplots for feature distributions by species...")
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='species', y='sepal_length')
plt.title("Sepal Length Distribution by Species")
plt.show()

sns.boxplot(data=df, x='species', y='sepal_width')
plt.title("Sepal Width Distribution by Species")
plt.show()

sns.boxplot(data=df, x='species', y='petal_length')
plt.title("Petal Length Distribution by Species")
plt.show()

sns.boxplot(data=df, x='species', y='petal_width')
plt.title("Petal Width Distribution by Species")
plt.show()
```
